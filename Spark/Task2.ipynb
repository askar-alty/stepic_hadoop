```python
import math
```

```python
servers = 50 * 12
```

```python
summm = []
for i in range(servers):
    for j in range(servers):
        if i + j == servers and i > j:
            summm.append((i, j))
            
```

```python
summm
```

```python
T = []
for (x, y) in summm:
    T.append((x, (20 * math.ceil(1000/x)) + (1000/y)))
```
